## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

sum = 0
exito = 0
num = 1000
for n in range(num):
    dados = random.randint(1, 6) + random.randint(1, 6)
    sum += dados
    if dados > 7 or dados % 2 == 0:
        exito += 1

print(f"{num} veces se tiraron los dados y {exito} veces fue más de 7 o un número par.")
print(f"En promedio el resultado fue: {sum/num}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

num = 10000
iguales = 0
whitered = 0
white = 0
red = 0
for x in range(num):
    b1 = ball_box[random.randint(0, 59)]
    b2 = ball_box[random.randint(0, 59)]
    b3 = ball_box[random.randint(0, 59)]
    b4 = ball_box[random.randint(0, 59)]
    b5 = ball_box[random.randint(0, 59)]
    if b1 == b2 and b3 == b2 and b4 == b3 and b4 == b5: iguales += 1
    if b1 == 'White': white +=1
    elif b1 == 'Red': red +=1
    if b2 == 'White': white +=1
    elif b2 == 'Red': red +=1
    if b3 == 'White': white +=1
    elif b3 == 'Red': red +=1
    if b4 == 'White': white +=1
    elif b4 == 'Red': red +=1
    if b5 == 'White': white +=1
    elif b5 == 'Red': red +=1
    if red == 2 and white == 3: whitered += 1
    white = 0
    red = 0

print(f"{num} veces se sacaron 5 bolas,")
print(f"{whitered} veces se sacaron 3 blancas y 2 rojas")
print(f"y {iguales} veces se sacaron todas del mismo color.")
```
